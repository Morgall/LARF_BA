```python
import pandas as pd
import time
import sys
from pathlib import Path
import numpy as np
from sklearn.tree import DecisionTreeClassifier


# Go up one directory to get to master/
project_root = str(Path.cwd().parent)
sys.path.append(project_root)
```

```python
from sklearn.model_selection import train_test_split

data_adult = pd.read_csv("../datasets/adult/stacked.csv")


features = data_adult.drop(columns=['y'])
targets = data_adult['y']

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.3, random_state=1, stratify=targets)

for depth in range(2, 8+1):
    # Initialize the Decision Tree Classifier
    clf = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=depth, random_state=1)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    df = pd.DataFrame([], columns=['y', 'prediction'])
    df['y'] = y_test
    df['prediction'] = y_pred
    print(df)




# Train the classifier
#clf.fit(X_train, y_train)

# Predict the classes for test data
#y_pred = clf.predict(X_test)

#print(y_pred)
```
